```python
import json
from contagion.utils import get_simluation_parameters
```

```python
def formated_contagion_parameters(params):
  return '\n'.join([
    '\t{',
    f'\t\t"tag": "{params["tag"]}",',
    f'\t\t"mu": {params["mu"]},',
    f'\t\t"betas": np.array([{params["beta1"]}]),',
    '\t\t"rho0s_per_beta_delta": {',
    f'\t\t\t{params["beta2"]}: {params["I_0"]}',
    '\t\t},',
    f'\t\t"timesteps": {params["timesteps"]},',
    f'\t\t"sigma": {params["sigma1"]},',
    f'\t\t"sigma_delta": {params["sigma2"]},',
    f'\t\t"bounded_delta": {params["bounded_delta"]},',
    f'\t\t"independent_noises": {params["independent_noises"]}',
    '\t},',
    '\t{',
    f'\t\t"tag": "{params["tag"]}_sigma0",',
    f'\t\t"mu": {params["mu"]},',
    f'\t\t"betas": np.array([{params["beta1"]}]),',
    '\t\t"rho0s_per_beta_delta": {',
    f'\t\t\t{params["beta2"]}: {params["I_0"]}',
    '\t\t},',
    f'\t\t"timesteps": {params["timesteps"]}',
    '\t}'
  ])
```

```python
dataset = 'InVS15'
bounded_delta = False
independent_noises = False

with open("HVHT22_contagion_experiments_by_region.json", encoding = 'utf-8') as f:
  region_parameters_list = json.load(f)

formated_list = []
for experiment in region_parameters_list:
  params = get_simluation_parameters(dataset=dataset, mu=experiment["parameters"]["mu"], sigma=experiment["parameters"]["sigma"], J0=experiment["parameters"]["J0"], J0_delta=experiment["parameters"]["J0_delta"])
  params["tag"] = f'J_{dataset}_{experiment["experiment"]}'
  params["I_0"] = experiment["parameters"]["I_0"]
  params["timesteps"] = experiment["parameters"]["timesteps"]
  params["bounded_delta"] = bounded_delta or False
  params["independent_noises"] = independent_noises or False
  
  print('\n'.join([
    f'tag = {params["tag"]}',
    f'\tk1 = {params["k1"]}',
    f'\tk2 = {params["k2"]}',
    f'\tmu = {params["mu"]}',
    f'\tbeta1 = {params["beta1"]}',
    f'\tsigma1 = {params["sigma1"]}',
    f'\tbeta2 = {params["beta2"]}',
    f'\tsigma2 = {params["sigma2"]}',
    f'\tbeta = {params["beta"]}',
    f'\tsigma = {params["sigma"]}',
    f'\tbeta_delta = {params["beta_delta"]}',
    f'\tsigma_delta = {params["sigma_delta"]}',
    f'\tJ0 = {params["J0"]}',
    f'\tJ0_delta = {params["J0_delta"]}',
    f'\tJ0_asterisk = {params["J0_asterisk"]}',
    f'\tR0 = {params["R0"]}',
    f'\tR0_delta = {params["R0_delta"]}',
    f'\tregion = {params["region"]}',
    f'\tXminus_sigma0 = {params["Xminus_sigma0"]}',
    f'\tI_0 = {params["I_0"]}',
    f'\ttimesteps = {params["timesteps"]}',
    f'\tbounded_delta = {params["bounded_delta"]}',
    f'\tindependent_noises = {params["independent_noises"]}',
    ''
  ]))
  formated_list.append(formated_contagion_parameters(params))

print('\n\nFormatted ouput for contagion_experiments.py:')
print(',\n'.join(formated_list))
```
